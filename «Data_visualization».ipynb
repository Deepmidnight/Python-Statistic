```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Deepmidnight/Midnight_nebula/main/Happiness_2019.csv')
```

```python
df.head()
```

```python
df.tail(10)
```

```python
print(df['Overall rank'].unique())
print(df['Country or region'].unique())
print(df['Score'].unique())
print(df['GDP per capita'].unique())
print(df['Social support'].unique())
print(df['Healthy life expectancy'].unique())
print(df['Freedom to make life choices'].unique())
print(df['Generosity'].unique())
print(df['Perceptions of corruption'].unique())
```

Построение графиков:

```python
#Топ-5 самых свободных стран
top10_freedom = df[['Country or region', 'Score']].sort_values('Score', ascending=False).head(10)
top10_freedom
```

```python
top10_freedom.plot(kind='bar')
plt.title('Топ 10 свободных стран по общему рейтингу')
plt.xticks(top10_freedom.index, top10_freedom['Country or region'], rotation=60, fontsize=12)
plt.ylabel('Score')
for i, val in enumerate(top10_freedom['Score'].values):
  plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'size':10})   
plt.show()
```

```python
#Зависимость между свободой делать жизненный выбор и общим уровнем жизни
df.plot(kind='scatter', 
        x='Freedom to make life choices', 
        y='Score', 
        s=df['Score']*10,
        title='Зависимость между свободой делать жизненный выбор и общим уровнем жизни')
# Из визуализации видно, что чем более свободны граждане в своём выборе, тем выше общий урвоень жизни в государстве 
```

```python
#Зависимость между восприятием коррупции и свободой делать жизненный выбор
sns.jointplot(x='Freedom to make life choices', y='Perceptions of corruption', data=df, kind='reg')
# Из визуализации виден очень ясный тренд, что чем больше у граждан свободы выбора, тем сильнее они реагируют на коррупцию
```

```python
df_plot = df[['Country or region', 'Perceptions of corruption']]
```

```python
df_plot.plot(title='График всприятия коррупции', 
                             xlabel='Country or region', 
                             ylabel='Perceptions of corruption')
# Из визуализации видно, что в более развитых и свободных странах на коррупцию реагируютна порядок активнее, чем в менее свободных. Отдельные пики в нижней половине списка имеют место быть, 
# т.к. в тоталитарных странах руководство государства будучи тотально коррумпированным, очень часто за коррупцию вводит наказанием смертную казнь, которая по своей сути вообще не решает 
# корневой проблемы, а лишь сильнее создаёт напряжённость в обществе и сжимает пружину народного гнева
```

```python
df_copy= df.copy()
df_copy['Social support'] = df_copy['Social support']*0.33
df_copy
```

```python
df_area = df_copy[['Country or region','Perceptions of corruption','Generosity', 'Social support']]
```

```python
# Зависимость щедрости населения, восприятия коррупции и социальной поддержки
df_area.plot(kind='area',
                             title='Зависимость щедрости населения от восприятия коррупции', 
                             xlabel='Country or region', 
                                stacked=False,
                             ylabel='Оценка показателя')
# Из визуализации виднт, что чем ниже реакция на коррупцию в государстве, тем ниже социальная поддержка со стороны государства (приоритет тратить деньги на граждан ниже, чем на оборонку,
# внутренние войска и охрану собственной власти) и тем выше щедрость населения, чаще всего обусловлена тем, что машина пропаганды, в том числе, давит на жалость через сбор на лечение
# тяжелобольных детей через SMS и прочие благотворительные фонды. В развитых же странах зачастую это просто не нужно, т.к. общество поддерживается субсидиями за счет государства
```

```python
# Зависимость между Здоровой продолжительностью жизни, свободой выбора и восприятием коррупции
sns.pairplot(df[['Healthy life expectancy',	'Freedom to make life choices',	'Perceptions of corruption']])
# Из визуализации видно, что там, где сильнее общество реагирует на коррупцию, выше продолжительность здоровой жизни и больше свободы выбора
# Отдельные выбросы в продолжительности жизни, имеют своё реальное место и часто обусловленыкультурными особенностями и географическим местоположением, 
# климатом наряду с экологией(например кавказские горцы и тот же китай, который направив вектор развития на диктатуру, уничтожает экологию в стране и создал тенденцию к снижению
# этого показателя. Или же Япония, которая имея множество нерешённых проблем с демократическими институтами, несмотря на технологическое равитие стагнировала в экономике в начале 2000х, 
# но тем не менее идёт в первых рядах в мире по вкладу экологические нормы и является лидером по количеству долгожителей в мире)
```
